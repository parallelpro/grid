# Observables and estimators
Before we dive ourself into the code, we need to understand how we usually estimate stellar parameters with a grid of stellar evolutionary models. Suppose we have four observables for each star, namely "Teff", "feh", "L" and "log_g". We also have a directory which stores the evolutionary models. For MIST (Choi et al. 2016, ApJ, 823:102) models, that means files which end with ".eep" (they are accessible from http://waps.cfa.harvard.edu/MIST/model_grids.html). Each file contains a track of simulation which evolves a star from birth to death. It presents the evolution of stellar parameters as a function of time. Each row (a time step) is regarded as a "model". By matching the observables with those associated with the model, we are able to find some decent representations of the observed star. They are further used to calculate any parameters, i.e. the estimators, we desire.

However, the Bayes' theorem is the popular way to quantify such problems.

-prior
-posterior


Okay, now let's look at the code. Four steps before we reach the summit.

# Step 1: define a function to read your tracks

The function takes one argument "trackfile" which is the filepath of an evolutionary track. The function reads the track, adds columns which are desired to use either as observables or estimators, and returns a table-like object (like the one defined in ``astropy``). The table should be able to extract one column value by calling the name of either observables or estimators. For example, ``table["Teff"]`` returns the effective temperature for the models of the current track.

```python
def read_mist_ms_tracks(trackfile):
    # ascii.read is a method provided by ``astropy`` to read text files. 
    table = ascii.read(trackfile, converters={'#':' '}, header_start=11)
    # only keep the main sequence part of a mist track
    table = table[table['phase'] == 0]

    # add observables in the table that do not exist: feh
    z_surf_solar, x_surf_solar = 0.017, 0.735 # gs98
    feh = np.log10((1.-table['surface_h1']-table['surface_he4']-
        table['surface_he3'])/table['surface_h1'])-np.log10(z_surf_solar/x_surf_solar)
    table['feh'] = feh

    # add observables that are power10 versions of the original columns
    cols = ['log_Teff', 'log_L', 'log_R']
    newcols = ['Teff', 'L', 'R']
    for icol in range(len(cols)):
        table[newcols[icol]] = 10.0**table[cols[icol]]

    return table
```

# Step 2: specify a list of tracks - your database
A simple search by ``os.walk`` would do.

```python
# search mist tracks
databasedir = '../dwarf_mist/'
for parentdir, dirname, filenames in os.walk(databasedir):
    a = [filenames[i].endswith('.track.eep') for i in range(0,len(filenames))]
    f = filenames
a, f = np.array(a), np.array(f)
tracks = f[a]
tracks = np.array([databasedir+tracks[itrack] for itrack in range(len(tracks))])
```

# Step 3: specify observables and estimators

```python
# output directory
outdir = 'output/'

# observables and estimators
observables = np.array(['Teff', 'feh', 'L', 'log_g'])
estimators = np.array(['star_mass', 'star_age', 'mass_conv_core', 
                            'Teff', 'L', 'R', 'log_g', 'delta_nu', 
                            'delta_Pg', 'nu_max'])

# A file containing the observables for my stars
stars = ascii.read('Mc14_table1_input_parameters.txt', format='csv', delimiter="|")[0:20]
stars_obs_qty = np.array([stars['teff_lamost_lasp'], 
                            stars['feh_lamost_lasp'], 
                            stars['Lum_gaia'], 
                            stars['logg_lamost_lasp']]).T
stars_obserr_qty = np.array([stars['teff_lamost_lasp_err'], 
                            stars['feh_lamost_lasp_err'], 
                            (stars['E_Lum']+stars['e_Lum'])/2.0, 
                            stars['logg_lamost_lasp_err']]).T
starnames = np.array(stars['kicnumber'], dtype='str')
```

# Step 4: run
We first initialize a grid class ``mygrid`` by passing values we defined in the previous step. Then run the estimation by calling ``estimate_parameters``. Use keyword ``Nthread`` to show your computer's muscle. 

You will be getting plot warnings (see below), but there's nothing to worry about.

```python
# initialize the grid class
mygrid=grid(read_mist_ms_tracks, tracks, outdir, 
    observables, estimators, stars_obs_qty, stars_obserr_qty, starname=starnames)

# run
mygrid.estimate_parameters(Nthread=40)
```
